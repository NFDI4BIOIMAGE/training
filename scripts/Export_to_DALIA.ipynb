```python
import re
from difflib import get_close_matches

# Return an http://spdx.org/licenses/ URI from a human-ish license string.
# Example: "bsd 3-clause"  -> "http://spdx.org/licenses/BSD-3-Clause.html"
#          "CC-BY-4.0 International" -> "http://spdx.org/licenses/CC-BY-4.0.html"

# A curated (not exhaustive) set of common SPDX IDs for fuzzy fallback.
SPDX_IDS = {
    # Most common
    "MIT","Apache-2.0","BSD-2-Clause","BSD-3-Clause","BSD-4-Clause","GPL-2.0-only",
    "GPL-2.0-or-later","GPL-3.0-only","GPL-3.0-or-later","LGPL-2.1-only","LGPL-2.1-or-later",
    "LGPL-3.0-only","LGPL-3.0-or-later","AGPL-3.0-only","AGPL-3.0-or-later","MPL-2.0",
    "EPL-1.0","EPL-2.0","CDDL-1.0","CDDL-1.1","MPL-1.1","IPL-1.0","NCSA","Zlib",
    "ISC","Unlicense","BSL-1.0","Artistic-2.0","WTFPL","OpenSSL","PostgreSQL","Python-2.0",
    "Ruby","Perl","OFL-1.1","OFL-1.1-RFN","Unicode-DFS-2016","CC0-1.0",
    "CC-BY-3.0","CC-BY-4.0","CC-BY-SA-3.0","CC-BY-SA-4.0","CC-BY-NC-4.0","CC-BY-NC-SA-4.0",
    "CC-BY-ND-4.0","CC-BY-NC-ND-4.0","EUPL-1.2","CPL-1.0","Apache-1.1","Apache-1.0",
    "BSD-3-Clause-Clear","BSD-3-Clause-No-Nuclear-License","BSD-3-Clause-No-Nuclear-Warranty",
    "BSD-Source-Code","HPND","NOSL","0BSD","EDL-1.0","CERN-OHL-P-2.0","CERN-OHL-S-2.0",
    "CERN-OHL-W-2.0","BlueOak-1.0.0","JSON","UPL-1.0","LGPL-2.0-only","LGPL-2.0-or-later",
    "GPL-1.0-only","GPL-1.0-or-later","AGPL-1.0-only","AGPL-1.0-or-later","AGPL-3.0-only",
}

# Regex-driven alias normalization to SPDX IDs
ALIAS_PATTERNS = [
    # MIT
    (re.compile(r"^\s*mit\s*$", re.I), "MIT"),
    (re.compile(r"\bexpat\b", re.I), "MIT"),

    # Apache 2
    (re.compile(r"^\s*apache(\s*license)?\s*(v|version)?\s*2(\.0)?\s*$", re.I), "Apache-2.0"),
    (re.compile(r"^\s*asl\s*2(\.0)?\s*$", re.I), "Apache-2.0"),

    # BSD (2/3/4)
    (re.compile(r"^\s*bsd\s*2(\s*clause|\s*-?\s*clause)?\s*$", re.I), "BSD-2-Clause"),
    (re.compile(r"^\s*bsd\s*3(\s*clause|\s*-?\s*clause)?\s*$", re.I), "BSD-3-Clause"),
    (re.compile(r"^\s*bsd\s*4(\s*clause|\s*-?\s*clause)?\s*$", re.I), "BSD-4-Clause"),
    (re.compile(r"^\s*bsd\s*clear\s*$", re.I), "BSD-3-Clause-Clear"),

    # ISC
    (re.compile(r"^\s*isc\s*$", re.I), "ISC"),

    # Zlib
    (re.compile(r"^\s*zlib\s*$", re.I), "Zlib"),

    # Boost
    (re.compile(r"^\s*boost\s*(software)?\s*license\s*(1\.0)?\s*$", re.I), "BSL-1.0"),

    # MPL/EPL/CDDL/IPL
    (re.compile(r"^\s*mpl\s*(v|version)?\s*2(\.0)?\s*$", re.I), "MPL-2.0"),
    (re.compile(r"^\s*epl\s*(v|version)?\s*2(\.0)?\s*$", re.I), "EPL-2.0"),
    (re.compile(r"^\s*epl\s*(v|version)?\s*1(\.0)?\s*$", re.I), "EPL-1.0"),
    (re.compile(r"^\s*cddl\s*(1\.1)?\s*$", re.I), "CDDL-1.1"),
    (re.compile(r"^\s*cddl\s*(1\.0)\s*$", re.I), "CDDL-1.0"),
    (re.compile(r"^\s*ipl\s*(1\.0)?\s*$", re.I), "IPL-1.0"),

    # GPL family (accept v, version, plus signs)
    (re.compile(r"^\s*gpl\s*(v|version)?\s*2(\.0)?\s*\+\s*$", re.I), "GPL-2.0-or-later"),
    (re.compile(r"^\s*gpl\s*(v|version)?\s*2(\.0)?\s*(-?only)?\s*$", re.I), "GPL-2.0-only"),
    (re.compile(r"^\s*gpl\s*(v|version)?\s*3(\.0)?\s*\+\s*$", re.I), "GPL-3.0-or-later"),
    (re.compile(r"^\s*gpl\s*(v|version)?\s*3(\.0)?\s*(-?only)?\s*$", re.I), "GPL-3.0-only"),
    (re.compile(r"^\s*gpl\s*(v|version)?\s*1(\.0)?\s*\+\s*$", re.I), "GPL-1.0-or-later"),
    (re.compile(r"^\s*gpl\s*(v|version)?\s*1(\.0)?\s*(-?only)?\s*$", re.I), "GPL-1.0-only"),

    # LGPL family
    (re.compile(r"^\s*lgpl\s*(v|version)?\s*2\.1\s*\+\s*$", re.I), "LGPL-2.1-or-later"),
    (re.compile(r"^\s*lgpl\s*(v|version)?\s*2\.1\s*(-?only)?\s*$", re.I), "LGPL-2.1-only"),
    (re.compile(r"^\s*lgpl\s*(v|version)?\s*2(\.0)?\s*\+\s*$", re.I), "LGPL-2.0-or-later"),
    (re.compile(r"^\s*lgpl\s*(v|version)?\s*2(\.0)?\s*(-?only)?\s*$", re.I), "LGPL-2.0-only"),
    (re.compile(r"^\s*lgpl\s*(v|version)?\s*3(\.0)?\s*\+\s*$", re.I), "LGPL-3.0-or-later"),
    (re.compile(r"^\s*lgpl\s*(v|version)?\s*3(\.0)?\s*(-?only)?\s*$", re.I), "LGPL-3.0-only"),

    # AGPL
    (re.compile(r"^\s*agpl\s*(v|version)?\s*3(\.0)?\s*\+\s*$", re.I), "AGPL-3.0-or-later"),
    (re.compile(r"^\s*agpl\s*(v|version)?\s*3(\.0)?\s*(-?only)?\s*$", re.I), "AGPL-3.0-only"),
    (re.compile(r"^\s*agpl\s*(v|version)?\s*1(\.0)?\s*\+\s*$", re.I), "AGPL-1.0-or-later"),
    (re.compile(r"^\s*agpl\s*(v|version)?\s*1(\.0)?\s*(-?only)?\s*$", re.I), "AGPL-1.0-only"),

    # Creative Commons (CC 4.0 defaults if version omitted)
    (re.compile(r"^\s*cc[-\s]?0(\.0)?\s*$", re.I), "CC0-1.0"),
    (re.compile(r"^\s*cc[-\s]?by\s*(4(\.0)?|four|v4|version\s*4(\.0)?)?\s*(international)?\s*$", re.I), "CC-BY-4.0"),
    (re.compile(r"^\s*cc[-\s]?by[-\s]?sa\s*(4(\.0)?|v4|version\s*4(\.0)?)?\s*$", re.I), "CC-BY-SA-4.0"),
    (re.compile(r"^\s*cc[-\s]?by[-\s]?nc\s*(4(\.0)?|v4|version\s*4(\.0)?)?\s*$", re.I), "CC-BY-NC-4.0"),
    (re.compile(r"^\s*cc[-\s]?by[-\s]?nc[-\s]?sa\s*(4(\.0)?|v4|version\s*4(\.0)?)?\s*$", re.I), "CC-BY-NC-SA-4.0"),
    (re.compile(r"^\s*cc[-\s]?by[-\s]?nd\s*(4(\.0)?|v4|version\s*4(\.0)?)?\s*$", re.I), "CC-BY-ND-4.0"),
    (re.compile(r"^\s*cc[-\s]?by[-\s]?nc[-\s]?nd\s*(4(\.0)?|v4|version\s*4(\.0)?)?\s*$", re.I), "CC-BY-NC-ND-4.0"),
    (re.compile(r"^\s*creative\s*commons\s*attribution\s*(4(\.0)?|international)?\s*$", re.I), "CC-BY-4.0"),
    (re.compile(r"^\s*creative\s*commons\s*attribution[-\s]?sharealike\s*(4(\.0)?|international)?\s*$", re.I), "CC-BY-SA-4.0"),

    # Others
    (re.compile(r"^\s*mozilla\s*public\s*license\s*(2(\.0)?)?\s*$", re.I), "MPL-2.0"),
    (re.compile(r"^\s*eclipse\s*public\s*license\s*2(\.0)?\s*$", re.I), "EPL-2.0"),
    (re.compile(r"^\s*unlicense\s*$", re.I), "Unlicense"),
    (re.compile(r"^\s*wtfpl\s*$", re.I), "WTFPL"),
    (re.compile(r"^\s*openssl\s*$", re.I), "OpenSSL"),
    (re.compile(r"^\s*postgresql\s*$", re.I), "PostgreSQL"),
    (re.compile(r"^\s*python\s*software\s*foundation\s*license\s*$", re.I), "Python-2.0"),
    (re.compile(r"^\s*ruby\s*license\s*$", re.I), "Ruby"),
    (re.compile(r"^\s*perl\s*license\s*$", re.I), "Perl"),
    (re.compile(r"^\s*upl\s*(1\.0)?\s*$", re.I), "UPL-1.0"),
    (re.compile(r"^\s*eupl\s*(1\.2)?\s*$", re.I), "EUPL-1.2"),
    (re.compile(r"^\s*ofl\s*(1\.1)?\s*$", re.I), "OFL-1.1"),
    (re.compile(r"^\s*0bsd\s*$", re.I), "0BSD"),
    (re.compile(r"^\s*bsd\s*source\s*code\s*$", re.I), "BSD-Source-Code"),
]

# Simple passthrough if the input already looks like a valid SPDX ID we know
def _looks_like_spdx_id(s: str) -> bool:
    s = s.strip()
    if s in SPDX_IDS:
        return True
    # case-insensitive check
    return s.upper() in {x.upper() for x in SPDX_IDS}

def to_spdx_id(s: str) -> str | None:
    """Best-effort normalization from a free-text license string to an SPDX ID."""
    if not s or not s.strip():
        return None
    raw = s.strip()

    # Direct hit?
    if _looks_like_spdx_id(raw):
        # Return canonical casing from our set if needed
        for id_ in SPDX_IDS:
            if id_.lower() == raw.lower():
                return id_
        return raw  # fallback

    # Normalize common symbols
    norm = re.sub(r"[\u2010-\u2015]", "-", raw)      # various dashes → hyphen
    norm = re.sub(r"\s+", " ", norm).strip()         # collapse whitespace

    # Try regex alias patterns
    for rx, spdx in ALIAS_PATTERNS:
        if rx.match(norm):
            return spdx

    # Heuristics for “GPL/LGPL/AGPL X (+)?”
    m = re.match(r"^\s*(a?g?l?)gpl\s*(?:v|version)?\s*(\d)(?:\.(\d))?\s*(\+|or\s*later|only)?\s*$",
                 norm, re.I)
    if m:
        which, major, minor, plus = m.group(1), m.group(2), m.group(3), m.group(4)
        fam = ("AGPL" if which.lower().startswith("ag") else
               "LGPL" if which.lower().startswith("lg") else "GPL")
        ver = f"{major}.{minor or '0'}"
        tail = "-or-later" if plus and plus.strip() in {"+", "or later"} else "-only"
        return f"{fam}-{ver}{tail}"

    # Heuristic for “CC-XYZ 4.0 …”
    m = re.match(r"^\s*cc[-\s]?([a-z-]+)\s*(\d(?:\.\d)?)?\s*", norm, re.I)
    if m:
        flavor = m.group(1).upper()
        ver = (m.group(2) or "4.0").replace(" ", "")
        # Normalize separators in flavor
        flavor = flavor.replace(" ", "").replace("_", "-")
        return f"CC-{flavor}-{ver}"

    # Fuzzy fallback against our SPDX_IDS set (very conservative)
    cleaned = re.sub(r"[^A-Za-z0-9.+-]", "", norm)
    candidates = get_close_matches(cleaned, SPDX_IDS, n=1, cutoff=0.90)
    if candidates:
        return candidates[0]

    return None

def to_spdx_uri(s: str, scheme: str = "http") -> str | None:
    """Return the http://spdx.org/licenses/<ID>.html URI, or None if unknown."""
    spdx_id = to_spdx_id(s)
    if not spdx_id:
        return None
    return f"{scheme}://spdx.org/licenses/{spdx_id}.html"


def license_to_spdx_uri(x):
    if isinstance(x, list):
        return [to_spdx_uri(str(x[0]))]
    else:
        return to_spdx_uri(str(x))


data["license"] = data["license"].apply(license_to_spdx_uri)
```

```python
def map_learning_resource(entry):
    # Skip empty or NaN rows
    if entry is None or (isinstance(entry, float) and pd.isna(entry)):
        return ""
    # Use a set to avoid duplicates
    matches = set()
    if isinstance(entry, list):
        for item in entry:
            if item in type_to_learning_resource:
                matches.add(type_to_learning_resource[item])
    elif entry in type_to_learning_resource:
        matches.add(type_to_learning_resource[entry])
    return " * ".join([m for m in matches if m is not None])

def map_media_type(entry):
    # Skip empty or NaN rows
    if entry is None or (isinstance(entry, float) and pd.isna(entry)):
        return ""
    # Use a set to avoid duplicates
    matches = set()
    if isinstance(entry, list):
        for item in entry:
            if item in type_to_media_type:
                matches.add(type_to_media_type[item])
    elif entry in type_to_media_type:
        matches.add(type_to_media_type[entry])
    return " * ".join([m for m in matches if m is not None])

# Apply the mapping functions
data["LearningResourceType"] = data["type"].apply(map_learning_resource)
data["MediaType"] = data["type"].apply(map_media_type)
```

```python
import pandas as pd
import re

def normalize_author_format(authors):
    # Helper function to reformat a single name
    def reformat_name(name):
        # Check if it's already in "Surname, Prename" format
        if "," in name:
            return name.strip()
        # If in "Prename Surname" format, convert to "Surname, Prename"
        parts = name.split()

        if len(parts) > 0:
            url = ""
            if parts[-1].startswith("http"):
                url = ", " + parts[-1]
                parts = parts[:-1]
            
            et_al = ['et', 'al.']
            if len(parts) == 2 and all(p not in et_al for p in parts):
                return f"{parts[1]}, {parts[0]}" + url
            if len(parts) == 3 and all(p not in et_al for p in parts):
                return f"{parts[2]}, {parts[0]} {parts[1]}" + url
        return name.strip()  # Return unchanged if not a simple name format


    # Convert single strings to lists for uniform processing
    if isinstance(authors, str):
        authors = [authors]
    authors = [str(a).strip() for a in authors]  # Ensure all elements are strings

    #print(authors)
    
    # Process each author entry
    formatted_authors = []
    for author in authors:
        formatted_authors.append(reformat_name(author))

    #print("A", formatted_authors)
    
    # Join all processed names with "*"
    return " * ".join(formatted_authors)

#normalize_author_format(["Robert Haase https://orcid.org/..."])
#normalize_author_format(["Wagner, Robert https://orcid.org/..."])

test = pd.DataFrame({"Authors":["Wagner, Robert https://orcid.org/..."]})
test["Authors"].apply(normalize_author_format)
```

```python
# Apply the normalization function
data["Authors"] = data["Authors"].apply(normalize_author_format)

data["Authors"]
```

## Link

```python
def urls_to_link(urls):
    if isinstance(urls, str):
        urls = [urls]
    link = None
    for url in urls:
        if url.startswith("https://doi"):
            link = url
            break
    if link is None:
        link = urls[0]
    return link
data["Link"] = data["url"].apply(urls_to_link)

data["Link"]
```

```python
def map_file_format(media_type, file_format):
    # If FileFormat already has a valid entry, return it as is
    if file_format is not None and file_format.strip() != "":
        return file_format
    # Map media types to specific file formats
    if media_type == "audio":
        return ".mp3"
    elif media_type == "video":
        return ".mp4"
    else:
        return ""  # Return empty string if no mapping is needed

# Apply the mapping function
data["FileFormat"] = data.apply(
    lambda row: map_file_format(row["MediaType"], row["FileFormat"]),
    axis=1
)
```
